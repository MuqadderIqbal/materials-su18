## NBA

```python
nba = Table.read_table('nba_20172018_salaries.csv')
nba
```

```python
# for each team, average salary paid for each position

nba.pivot('Team', 'Position', 'Salary', np.average)
```

```python
# for each team,
# amount paid to "starter" (player earning the most) in each position

starters = nba.pivot('Position', 'Team', 'Salary', max)
starters 
```

```python
# sort teams by total amount paid to starters

total = starters.drop(0).apply(sum)
starters.with_column('Total', total).sort('Total', descending = True)
```

## Table methods

```python
drinks = Table(['Drink', 'Cafe', 'Price']).with_rows([
    ['Milk Tea', 'Tea One', 4],
    ['Espresso', 'Nefeli',  2],
    ['Coffee',    'Nefeli', 3],
    ['Espresso', "Abe's",   2]
])
drinks
```

```python
discounts = Table().with_columns(
    'Coupon % off', make_array(5, 50, 25),
    'Location', make_array('Tea One', 'Nefeli', 'Tea One')
)
discounts
```

```python
joined = drinks.join('Cafe',discounts,'Location')
joined
```

```python
joined_new = joined.with_column('Discounted Price', joined.column('Price')* (1 - joined.column(3)/100)).drop(2,3)
joined_new
```

```python
joined_new.sort('Discounted Price')
```

```python
joined_new.sort('Discounted Price').sort('Cafe', distinct = True)
```

```python
joined_new.group('Cafe', min)
```
